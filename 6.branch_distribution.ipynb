# 지점별 반품량

## 1년간 지점별 총 반품량 및 반품소요시간의 평균

```python
import matplotlib.pyplot as plt

# 1. 지점별 반품량 계산
A_returns = martA['반품일자'].notnull().sum()
B_returns = martB['반품일자'].notnull().sum()
C_returns = martC['반품일자'].notnull().sum()
D_returns = martD['반품일자'].notnull().sum()

# 2. 지점별 평균 반품 소요기간 계산
A_avg_return_duration = martA[martA['반품소요기간'] > 0]['반품소요기간'].mean()
B_avg_return_duration = martB[martB['반품소요기간'] > 0]['반품소요기간'].mean()
C_avg_return_duration = martC[martC['반품소요기간'] > 0]['반품소요기간'].mean()
D_avg_return_duration = martD[martD['반품소요기간'] > 0]['반품소요기간'].mean()

# 시각화
marts = ['Mart A', 'Mart B', 'Mart C', 'Mart D']
returns = [A_returns, B_returns, C_returns, D_returns]
avg_durations = [A_avg_return_duration, B_avg_return_duration, C_avg_return_duration, D_avg_return_duration]

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'mediumseagreen'
ax1.bar(marts, returns, color=color)
ax1.set_ylabel('반품량', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.plot(marts, avg_durations, color=color, marker='o')
ax2.set_ylabel('평균반품소요기간', color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('지점별 반품량 및 평균반품소요기간')
plt.show()
```

- 지점별 반품량은 C지점이 가장 높음
- 그 다음으로는 D지점이 높았다.
- 판매량 대비 반품량을 구해보면 좋을 듯
- 평균 반품 소요기간은 mart D지점이 1.7일로 가장 길었고 mart B가 가장 짧았음
- 평균 반품 소요기간의 가장 낮고 높은 값의 차이가 약 0.7-0.8차이라 의미가 있을지 ..

## 1년간 지점별 판매량 대비 반품량

```python
import matplotlib.pyplot as plt

# D마트에서 반품한 경우만 선택
returns_D = martD[~martD['반품일자'].isnull()]

# 반품 물품의 물품 소분류별 개수 계산
return_item_counts = returns_D['물품소분류'].value_counts()

# 상위 10개의 물품 소분류 선택
top_10_return_items = return_item_counts.head(10)

# 파이차트 생성
plt.figure(figsize=(10, 8))
top_10_return_items.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("D마트의 반품 물품 소분류 비율 (Top 10)")
plt.axis('equal')  # 원을 동그랗게 표시
plt.show()
```

```python
# 각 지점에서의 총 판매량 계산
A_total_sales = martA['구매수량'].sum()
B_total_sales = martB['구매수량'].sum()
C_total_sales = martC['구매수량'].sum()
D_total_sales = martD['구매수량'].sum()

# 각 지점에서의 총 반품량 계산
A_total_returns = martA['반품일자'].notnull().sum()
B_total_returns = martB['반품일자'].notnull().sum()
C_total_returns = martC['반품일자'].notnull().sum()
D_total_returns = martD['반품일자'].notnull().sum()

# 판매량 대비 반품량 계산
A_returns_ratio = A_total_returns / A_total_sales
B_returns_ratio = B_total_returns / B_total_sales
C_returns_ratio = C_total_returns / C_total_sales
D_returns_ratio = D_total_returns / D_total_sales

# 막대 그래프로 표시
labels = ['Mart A', 'Mart B', 'Mart C', 'Mart D']
ratios = [A_returns_ratio, B_returns_ratio, C_returns_ratio, D_returns_ratio]

plt.figure(figsize=(10, 6))
plt.bar(labels, ratios, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('지점별 판매량 대비 반품량')
plt.xlabel('Mart')
plt.ylabel('반품량/판매량')
plt.show()
```

```python
# 반품을 한 경우만 선택
import matplotlib.pyplot as plt

def find_most_returned_item(df):
    # 반품을 한 경우만 선택
    returns = df[(df['반품소요기간'] > 0) | (~df['반품일자'].isnull())]

    # 각 물품소분류별로 반품 횟수를 세어줌
    most_returned_item = returns['물품소분류'].value_counts().idxmax()

    return most_returned_item

# 각 마트에서 최다 반품 물품소분류 찾기
most_returned_items = {}
for mart_name, mart_df in [('Mart A', martA), ('Mart B', martB), ('Mart C', martC), ('Mart D', martD)]:
    most_returned_items[mart_name] = find_most_returned_item(mart_df)

print("최다 반품 물품소분류:")
for mart, item in most_returned_items.items():
    print(f"{mart}: {item}")
```

```python
# 구매월을 기준으로 데이터를 그룹화하여 각 그룹 내에서 연령대 별로 합계를 계산합니다.
total_purchase_by_month_age = combined_data.groupby(['구매월', '연령대'])['구매금액'].sum().reset_index()

# 각 연령대별로 해당 달의 평균 구매금액을 계산합니다.
average_purchase_by_month_age = total_purchase_by_month_age.groupby('연령대').apply(lambda x: x.groupby('구매월')['구매금액'].mean()).reset_index()

# 결과 출력
print(average_purchase_by_month_age)
```

```python
import matplotlib.pyplot as plt

# 막대 그래프 생성
plt.figure(figsize=(10,6))

# 각 연령대별로 막대 그래프를 생성합니다.
for age_group in average_purchase_by_month_age['연령대'].unique():
    data = average_purchase_by_month_age[(average_purchase_by_month_age['연령대'] == age_group) & (average_purchase_by_month_age['구매월'] != 11)]
    plt.bar(data['구매월'], data['구매금액'], label=f'연령대 {age_group}')

# 그래프 제목 및 축 레이블 설정
plt.title('연령대별 월별 평균 구매금액')
plt.xlabel('구매월')
plt.ylabel('평균 구매금액')

# 범례 추가
plt.legend()

# 색상 설정
plt.gca().set_prop_cycle(None) # 색상 초기화
colors = ['mediumseagreen'] * len(average_purchase_by_month_age['연령대'].unique())
plt.gca

# 그래프 출력
plt.tight_layout()
plt.show()
```
