## 각 컬럼 고유값(범주) 확인

```python
# 컬럼 타입에 따라 정보 출력
for column in mem_info.columns:
    print(f"{column} 컬럼:")
    if column == 'member_id':
        print(mem_info[column].nunique())
    elif mem_info[column].dtype == 'object':
        print(mem_info[column].value_counts())
    else:
        print(mem_info[column].describe())
    print()
```

```python
# 가입 회원 수
print("회원번호 컬럼:")
print(mem_info['member_id'].nunique())  # 회원번호는 고유값이므로 개별 값의 빈도를 출력
```

# 각 컬럼 분포(파이차트) 확인

```python
import matplotlib as mpl

plt.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 각 컬럼에 대해 개별적으로 파이차트 그리기
categorical_columns = [col for col in mem_info.columns if mem_info[col].dtype == 'object']

for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    plt.title(f"Distribution of {column}")

    # 각 범주의 빈도 계산하여 파이차트 그리기
    mem_info[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
    plt.ylabel('')
    plt.show()
```
