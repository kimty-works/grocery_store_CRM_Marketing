# 배송서비스여부, 모바일알람여부 보기

귀무가설(H0; Null Hypothesis): 두 범주형 변수 간에는 관련성이 없다. 즉, 변수 간에는 차이가 없다.

대립가설(H1; Alternative Hypothesis): 두 범주형 변수 간에는 관련성이 있다. 즉, 변수 간에는 차이가 있다.

```python
import matplotlib.pyplot as plt

# 각 회원번호를 기준으로 중복을 제거한 후, 배송서비스신청여부와 모바일알람여부의 빈도 계산
unique_customers = dlv.drop_duplicates(subset=['회원번호'])
delivery_service_counts = unique_customers['배송서비스신청여부'].value_counts()
mobile_alarm_counts = unique_customers['모바일알람여부'].value_counts()

# 파이차트의 색상 설정
colors = ['#FF9999', '#66B2FF']  # 예쁜 색상으로 설정

# 파이차트 그리기
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 배송서비스신청여부 파이차트
axes[0].pie(delivery_service_counts, labels=delivery_service_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
axes[0].set_title('Delivery Service Subscription')

# 모바일알람여부 파이차트
axes[1].pie(mobile_alarm_counts, labels=mobile_alarm_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
axes[1].set_title('Mobile Alarm Subscription')

plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 중복을 제거한 후, 각 등급에 대해 배송 서비스 신청 여부와 모바일 알람 여부의 빈도 계산
unique_customers = dlv.drop_duplicates(subset=['회원번호'])

# 배송 서비스 신청 여부와 모바일 알람 여부의 빈도를 카이제곱 검정을 위해 crosstab으로 만듦
crosstab_result = pd.crosstab(unique_customers['Grade'], [unique_customers['모바일알람여부']])

# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency(crosstab_result)

print("Chi-Square Test Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
from scipy.stats import shapiro

# 중복을 제거한 후, 각 등급에 대해 배송 서비스 신청 여부의 빈도 계산
unique_customers = dlv.drop_duplicates(subset=['회원번호'])

# 등급별로 배송 서비스 신청이 '신청'인 비율 계산
grade_service_counts = unique_customers.groupby('Grade')['배송서비스신청여부'].value_counts().unstack()

# 각 등급에서 배송 서비스 신청이 '신청'인 사람의 비율 계산
grade_service_ratio = grade_service_counts['신청'] / (grade_service_counts['신청'] + grade_service_counts['미신청'])

# 각 등급별로 배송 서비스 신청 비율을 리스트로 저장
grade_service_ratio_lists = [grade_service_ratio.loc[grade] for grade in grade_service_ratio.index]


# Shapiro-Wilk test를 사용하여 각 등급별 신청 비율의 정규성 검정 수행
for i, ratio_list in enumerate(grade_service_ratio_lists):
    stat, p = shapiro(ratio_list)
    print(f"Grade {i+1} - Shapiro-Wilk Statistic: {stat}, p-value: {p}")
```

```python
# 예제 데이터
example_data = [0.09476490547746001, 0.010913799482056974, 0.4269090909090909, 0.009259259259259259]

# 샤피로-윌크 정규성 검정 수행
stat, p = shapiro(example_data)
print("Shapiro-Wilk Statistic:", stat)
print("p-value:", p)
```

```python
from scipy.stats import kruskal

# Kruskal-Wallis H 검정 수행
h_statistic, p_value = kruskal(*grade_service_ratio_lists)

print("Kruskal-Wallis H Statistic:", h_statistic)
print("P-value:", p_value)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 중복을 제거한 후, 각 등급에 대해 알람 서비스 신청 여부의 빈도 계산
unique_customers = dlv.drop_duplicates(subset=['회원번호'])

# 등급별로 알람 서비스 신청이 '신청'인 비율 계산
grade_service_counts = unique_customers.groupby('Grade')['모바일알람여부'].value_counts(normalize=True).unstack()

# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency(grade_service_counts)

print("Chi-Square Test Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```
