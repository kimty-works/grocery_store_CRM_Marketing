# 멤버 데이터 전처리

## 기타-00동인 데이터: 기타(구) 채워넣기

```python
# 변경 사전 만들기
replace_dict = {
    "기타-금광동": "중원구-금광동",
    "기타-신흥동": "수정구-신흥동",
    "기타-금곡동": "분당구-금곡동",
    "기타-동천동": "수지구-동천동",
    "기타-중앙동": "처인구-중앙동",
    "기타-보라동": "기흥구-보라동",
    "기타-동백동": "기흥구-동백동",
    "기타-은행동": "중원구-은행동",
    "기타-궁내동": "분당구-궁내동",
    "기타-율동": "분당구-율동",
    "기타-태평동": "수정구-태평동",
    "기타-성남동": "중원구-성남동",
    "서울 강남구-은행동": "중원구-은행동"
}

# 변경된 주소값으로 '주소' 열 값 변경
member['주소'] = member['주소'].replace(replace_dict)
```

## 멤버 데이터 null값 대체

```python
member.isnull().sum()
```

```python
# '성별' 열의 결측치를 최빈값으로 대체
mode_gender = member['성별'].mode()[0]
member['성별'] = member['성별'].fillna(mode_gender)

# '연령' 열의 결측치를 최빈값으로 대체
mode_age = member['연령'].mode()[0]
member['연령'] = member['연령'].fillna(mode_age)
```

```python
# 결측치 재확인
member.isnull().sum()
```

# 시 파생변수 생성

```python
# 시로 파생변수 생성
# '시'열을 추가
member['시'] = ''

# '주소'열을 기반으로 '시'열 값 설정
for index, row in member.iterrows():
    구, 동 = row['주소'].split('-')
    구 = 구.split()[-1]
    if 구 in ['영통구', '팔달구', '장안구', '권선구']:
        member.at[index, '시'] = '수원'
    elif 구 in ['수지구', '처인구', '기흥구']:
        member.at[index, '시'] = '용인'
    elif 구 in ['수정구', '분당구', '중원구']:
        member.at[index, '시'] = '성남'
    elif 구 == '광주':
        member.at[index, '시'] = '광주'
    elif 구 == '하남':
        member.at[index, '시'] = '하남'
    elif 구 == '화성':
        member.at[index, '시'] = '화성'
    elif 구 in ['강남구', '송파구']:
        member.at[index, '시'] = '서울'
    elif 구 in ['기타']:
        member.at[index, '시'] = '기타'


# '시'열을 주소 열 왼쪽에 삽입
member.insert(2, '시', member.pop('시'))

# '시' 열에 결측값 있는지 확인
member.isnull().sum()
```

```python
# '주소'열을 기반으로 구와 동 정보만 남기기
for index, row in member.iterrows():
    split_address = row['주소'].split('-')
    # 시 정보가 포함된 경우 시 정보를 제거한 후, 구와 동 정보만 남기기
    split_address = split_address[0].split()
    if len(split_address) == 2:
        member.at[index, '주소'] = split_address[-1]
```

# 연령대 파생변수 생성

```python
# 20대 이하/30대/40대/50대/60대/70대/80대 이상
def get_age_group(age):
    if age <= 20:
        return '20대 이하'
    elif age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    elif age < 50:
        return '40대'
    elif age < 60:
        return '50대'
    elif age < 70:
        return '60대'
    elif age < 80:
        return '70대'
    else:
        return '80대 이상'

# '연령' 열을 기반으로 '연령대' 열을 추가
member['연령대'] = member['연령'].apply(get_age_group)

# '연령대'열을 주소 열 왼쪽에 삽입
member.insert(6, '연령대', member.pop('연령대'))

# 결과 출력
member
```
